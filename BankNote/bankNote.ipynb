```python
import pandas as pd

# importing keras from tensorflow
from tensorflow import keras
# Import confusion matrix & classification report in order to evaluate the model
from sklearn.metrics import confusion_matrix , classification_report

# Import seaborn for graph plotting
import seaborn as sns


bank_note_data  = pd.read_csv("https://raw.githubusercontent.com/dphi-official/Datasets/master/bank_note_data/training_set_label.csv" )
```

```python
import numpy as np
```

```python
bank_note_data.head()
```

```python
bank_note_data['Class'].value_counts()
```

```python
X = bank_note_data.drop('Class',axis=1)
```

```python
y = bank_note_data['Class']
```

```python
from sklearn.model_selection import train_test_split
```

```python

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=15, stratify=y)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
# BUILDING A DEEP LEARNING MODEL

# Importing tensorflow as tf
import tensorflow as tf
# importing keras from tensorflow
from tensorflow import keras



model = keras.Sequential([
# first layer which has 12 input layer & activation is relu
keras.layers.Dense(4, input_dim=4, activation='relu'),
# First hidden layer
keras.layers.Dense(2, activation='relu'),
# Second hidden layer
keras.layers.Dense(1, activation='sigmoid')
])

# since it is binary classification metrics is accuracy & loss is binary_crossentropy
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
# Fitting the model X_train & y_train & running for 100 times(100 epochs)
model.fit(X_train, y_train, epochs=500)
```

```python
print(model.evaluate(X_test, y_test))
# Accuracy comes out to be 75.21 %
```

```python
# ----------Prediction on the test data------------------
yp = model.predict(X_test)
yp[:5]
```

```python
# Simple program for if predicted value is greater than 0.5 then 1 else 0
y_pred = []
for element in yp:
  if element > 0.5:
    y_pred.append(1)
else:
  y_pred.append(0)
```

```python
test_data = pd.read_csv('https://raw.githubusercontent.com/dphi-official/Datasets/master/bank_note_data/testing_set_label.csv')
```
