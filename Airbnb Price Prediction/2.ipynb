```python
import pandas as pd
```

```python
result = pd.read_csv('/content/train.csv')
```

```python
result2 = result.drop(['Unnamed: 0'],axis=1)
```

```python
result2['Bedrooms'].fillna(result2['Bedrooms'].mean())

# result2['Bedrooms'].fillna(data2['Bedrooms'].mean())

mean_value=result2['Bedrooms'].median()
  
result2['Bedrooms'].fillna(value=mean_value, inplace=True)
```

```python
result2
```

```python
X = result2.drop(['Price'],axis=1)
```

```python
y = result2['Price']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=.10,random_state =123)
```

```python
from sklearn.neural_network import MLPRegressor
```

```python

regr = MLPRegressor()
```

```python
regr.fit(X_train,y_train)
```

```python
X_train
```

```python
pred = regr.predict(X_test)
```

```python
pred
```

```python
test_data = pd.read_csv('/content/test.csv')
```

```python
test_data2 = test_data.drop(['Unnamed: 0'],axis=1)
```

```python
target = regr.predict(test_data2)
```
